```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import uproot
%matplotlib inline
```

```python
samples = uproot.open("config1/samples_beam_1_ip_150.root")['transport_samples']
samples.keys()
```

```python
df = samples.pandas.df([key for key in samples.allkeys() if len(samples[key].array()) > 0])
df
```

```python
plt.plot(df['x_in'], df['theta_x_in'], 'ro')
```

```python
plt.plot(df['x_in'], df['y_in'], 'ro')
```

```python
plt.plot(df['x_in'], df['def_x_out'], 'ro')
```

```python
# now w/o matplotlib, only with pandas builtins
df.plot(kind = 'scatter', x = 'x_in', y = 'y_in')
```

```python
# now lets try adding some fancy stuff
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
df.plot(kind = 'scatter', x = 'x_in', y = 'y_in', ax=ax)
ax.set_title('Scatter plot of y_in vs x_in')
ax.set_xlabel('x_in')
ax.set_ylabel('y_in')
```

```python
# bar plots, but theres no categorical data, so lets try histograms
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
df['x_in'].plot.hist(ax=ax)
ax.set_title('Histogram of x_in')
ax.set_xlabel('x_in')
ax.set_ylabel('Number of elements')
```

```python
# would try box plots but again, no categorical data
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
df[['x_in', 'y_in']].boxplot(ax = ax)
ax.set_title('Boxplot of x_in and y_in without grouping bc no categorical data')
ax.set_xlabel('x_in')
ax.set_ylabel('y_in')
```

```python
# now some seaborn
sns.set_style('whitegrid')
sns.kdeplot(df['x_in'])
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
sns.set_style('whitegrid')
sns.kdeplot(df['x_in'], ax = ax)
ax.set_title('KDE plot of x_in')
ax.set_xlabel('x_in')
ax.set_ylabel('density')
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
sns.set_style('whitegrid')
sns.kdeplot(df[['x_in', 'y_in']], ax = ax, cmap='Blues_d')
ax.set_title('KDE plot of x_in and y_in')
ax.set_xlabel('x_in')
ax.set_ylabel('y_in')
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
sns.set_style('whitegrid')
sns.violinplot(x = 'x_in', y = 'ksi_in', data = df, ax = ax)
ax.set_title('Violin plot of x_in and ksi_in')
ax.set_xlabel('x_in')
ax.set_ylabel('ksi_in')
# not really working, should get two nice violins :(
```

```python
sns.lmplot(x = 'x_in', y = 'ksi_in', data = df, palette='Set2', fit_reg=True)
```

```python
sns.pairplot(df[['x_in', 'theta_x_in', 'y_in', 'theta_y_in']], palette='Set2', diag_kind='kde', size=2).map_upper(sns.kdeplot, cmap='Blues_d')
```

```python
g = sns.FacetGrid(df, col='x_in', row='theta_y_in')
g.map(sns.regplot, 'y_in', 'ksi_in', fit_reg=False)
```
