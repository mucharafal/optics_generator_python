```python
x = grid.T[0]
y = grid.T[5]
frame = pd.DataFrame(data={"x_in [m]": x, "dispersion_from_x [m]": y})
sns.scatterplot(x="x_in [m]", y="dispersion_from_x [m]",
                data=frame)
```

```python
sns.jointplot(x, y, kind="kde", height=5, space=0).set_axis_labels(xlabel="x in [m]", ylabel="dispersion_from_x [m]").fig.suptitle("Dispersion in depend from input x")
```
